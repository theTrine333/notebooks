```python
import requests
from bs4 import BeautifulSoup
```

```python
main_url = "https://ww4.fmovies.co/movies/"
```

```python
req = requests.get(main_url)
print(req)
```

```python
soup = BeautifulSoup(req.content,"html.parser")

def getMovies():
    movies = []
    main_container = soup.find("div",class_='card-body p-0')
    movies_list = main_container.findAll('div',class_='col')
    
    for item in movies_list:
        link = item.find("a")['href']
        title = item.find("h2").getText()
        poster = item.find("img")['data-src']
        quality = item.find("span",class_='mlbq').getText()

        movies.append({
            'title':title,
            'link':link,
            'poster':poster,
            'quality':quality
        }
    )
    return movies
        

getMovies()
```
